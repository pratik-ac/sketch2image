```python
import cv2
import matplotlib.pyplot as plt
import os
import matplotlib
from tqdm.notebook import tqdm
from multiprocessing import Pool
from multiprocessing import cpu_count
```

```python
image_path = './Nepceleb'
sketch_path = './Nepceleb_sketch'
```

```python
def burnV2(image, mask):
    return 255 - cv2.divide(255-image, 255-mask, scale=256)
```

```python
def dodgeV2(image, mask):
    return cv2.divide(image, 255-mask, scale=256)
```

```python
def img_to_sketch(img_name):  # Change to accept a single image name
    img_rgb = cv2.imread(os.path.join(image_path, img_name))  # Use img_name here
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
    img_gray_inv = 255 - img_gray
    img_blur = cv2.GaussianBlur(img_gray_inv, ksize=(21, 21), sigmaX=0, sigmaY=0)
    img_blend = dodgeV2(img_gray, img_blur)
    matplotlib.image.imsave(os.path.join(sketch_path, img_name), img_blend, cmap='gray')
        
        
```

```python
img_names = os.listdir(image_path)
```

```python

print("[INFO] launching pool using {} processes...".format(8))
with Pool(processes=8) as pool:  # Use a context manager for the Pool
    # Use tqdm to wrap pool.imap for progress tracking
    for _ in tqdm(pool.imap(img_to_sketch, img_names), total=len(img_names), desc='Processing Images'):
        pass  # This will show progress with tqdm

print("[INFO] multiprocessing complete")
```

```python
# Display the last processed image 
if img_names:  # Check if there are images to display
    img_rgb = cv2.imread(os.path.join(image_path, img_names[-1]))  # Read the last image as an example
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
    img_gray_inv = 255 - img_gray
    img_blur = cv2.GaussianBlur(img_gray_inv, ksize=(21, 21), sigmaX=0, sigmaY=0)
    img_blend = dodgeV2(img_gray, img_blur)
    plt.imshow(img_blend, cmap='gray')
    plt.show()
```
